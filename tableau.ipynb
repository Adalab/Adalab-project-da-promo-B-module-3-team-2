```python
df = pd.read_csv("datos/limpieza_tableau.csv", index_col =0)
df
```

```python
df.season.unique()
```

```python
season_cambio2 = {1:"primavera", 2:"verano", 3:"otoño", 4:"invierno"}
```

```python
df["season"] = df["season"].astype("object")
```

```python
df["season"]=df["season"].map(season_cambio2)
df
```

```python
semana_cambio = {1:"lunes", 2:"martes", 3:"miércoles",4:"jueves", 5:"viernes", 6:"sábado", 7:"domingo"}
```

```python
df["weekday"] = df["weekday"].astype("object")
```

```python
df["weekday"] = df["weekday"].map(semana_cambio)
```

```python
df
```

```python
laborable = {0:"no laborable", 1:"laborable"}
```

```python
df["workingday"] = df["workingday"].astype("object")
```

```python
df["workingday"] = df["workingday"].map(laborable)
df
```
